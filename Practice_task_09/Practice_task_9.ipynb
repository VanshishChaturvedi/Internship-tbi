# **Problem Statement**
You are given a sales dataset for a retail chain. Your task is to use Pandas groupby and aggregation to produce a region-level and category-level summary.

**Dataset** <br>
Create the following DataFrame in your notebook:
 ``` python
import pandas as pd

data = {
    'sale_id':   [1, 2, 3, 4, 5, 6, 7, 8],
    'region':    ['North', 'South', 'North', 'East', 'South', 'East', 'North', 'South'],
    'category':  ['Electronics', 'Clothing', 'Clothing', 'Electronics', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'amount':    [1500, 800, 600, 2200, 950, 700, 1800, 450]
}
df = pd.DataFrame(data)
 ```

**Tasks:**<br>
Compute total sales per region: Group by region, compute the sum of amount, reset the index, and rename the amount column to total_sales.<br><br>
Compute average sales per category: Group by category, compute the mean of amount, reset the index, and rename the amount column to avg_sales.<br><br>
Multi-column grouping: Group by both region and category simultaneously, compute the total amount using .agg(), reset the index, and rename the aggregated column to regional_category_total.<br><br>
Print all three resulting DataFrames.

````python
import pandas as pd

data = {
    'sale_id':   [1, 2, 3, 4, 5, 6, 7, 8],
    'region':    ['North', 'South', 'North', 'East', 'South', 'East', 'North', 'South'],
    'category':  ['Electronics', 'Clothing', 'Clothing', 'Electronics', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'amount':    [1500, 800, 600, 2200, 950, 700, 1800, 450]
}
df = pd.DataFrame(data)
````

````python
df
````

````python
df.groupby('region').agg(total_sales=('amount','sum')).reset_index()
````

````python
df.groupby('category').agg(avg_sales=('amount','mean')).reset_index()
````

````python
df.groupby(['region', 'category']).agg(regional_category_total=('amount','sum')).reset_index()
````

## One Block Solution

``` python
import pandas as pd

data = {
    'sale_id':   [1, 2, 3, 4, 5, 6, 7, 8],
    'region':    ['North', 'South', 'North', 'East', 'South', 'East', 'North', 'South'],
    'category':  ['Electronics', 'Clothing', 'Clothing', 'Electronics', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'amount':    [1500, 800, 600, 2200, 950, 700, 1800, 450]
}
df = pd.DataFrame(data)
df.groupby('region').agg(total_sales=('amount','sum')).reset_index()
df.groupby('category').agg(avg_sales=('amount','mean')).reset_index()
df.groupby(['region', 'category']).agg(regional_category_total=('amount','sum')).reset_index()
```
