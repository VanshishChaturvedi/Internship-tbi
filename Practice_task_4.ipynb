# **Problem Statement:**

A payroll analyst at a small company needs a Python module that processes employee salary data. The module must provide three reusable functions that compose together:

**count_taxed(employees)** — counts how many employees earn above a monthly tax threshold and returns that count as an integer.<br><br>
**total_salary(salaries)** — sums a list of numeric salary values and returns the total.<br><br>
**average_salary(salaries)** — computes the integer average (using //) by calling total_salary internally rather than re-implementing the summation.<br>

After defining the three functions, call each one with the sample data provided and print the labelled results.


## **Constraints & Requirements:**

Use def and return inside every function body — do not use print inside any function.<br><br>
count_taxed must unpack each employee tuple as name, salary inside the loop and use += to accumulate the count.<br><br>
average_salary must call total_salary internally — do not rewrite the summation.<br><br>
Define the monthly tax threshold as MONTHLY_TAX_THRESHOLD = 600000 / 12 as a constant outside all function bodies.<br><br>
Use += shorthand for accumulation inside total_salary.<br><br>
Use len() to obtain the number of salaries inside average_salary.<br><br>
No external libraries — Python built-ins only.


**Sample data:**

employees = [ <br>
    ("Alice",   80000),<br>
    ("Bob",    200000),<br>
    ("Carol",   45000),<br>
    ("Dave",   120000),<br>
    ("Eve",     30000),<br>
]<br>

salaries = [80000, 200000, 45000, 120000, 30000]

```python
MONTHLY_TAX_THRESHOLD = 600000 / 12
def count_taxed(employees):
  count = 0
  for employee in employees:
    if employee[1] >= MONTHLY_TAX_THRESHOLD:
      count += 1
  return count
def total_salary(salaries):
  total = 0
  for salary in salaries:
    total += salary
  return total
def average_salary(salaries):
  return total_salary(salaries) // len(salaries)
employees = [
("Alice", 80000),
("Bob", 200000),
("Carol", 45000),
("Dave", 120000),
("Eve", 30000),
]

salaries = [80000, 200000, 45000, 120000, 30000]
print("employees earn above a monthly tax threshold: using count_taxed(employees) -->")
print(count_taxed(employees))
print("sums a list of numeric salary values: using total_salary(salaries) -->")
print(total_salary(salaries))
print("computes the integer average of salaries: using average_salary(salaries) -->")
print(average_salary(salaries))
```
