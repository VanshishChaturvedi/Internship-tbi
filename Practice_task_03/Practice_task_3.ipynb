# **Problem Statement:**

A data analyst at an online education platform is building a report to identify high-performing students across departments. <br><br> The platform stores data in three separate tables: a student registration table, a marks table, and a department table. <br><br>The analyst must merge all three tables, filter to students who scored above 80, and then compute the average score per department using a chained merge workflow.


**Constraints & Requirements:**

Use pandas only (import pandas as pd); no other libraries are required.<br><br>
Merge the three DataFrames using chained pd.merge() calls.<br><br>
Use an inner join when merging the student table with the marks table (only students with scores should appear).<br><br>
Use a left join when merging the intermediate result with the department table (preserve all matched students even if a department record is somehow missing).<br><br>
After merging, filter the result to rows where marks > 80.<br><br>
Compute the average marks per dept_name using .groupby().<br><br>
Handle the case where the marks and department tables use different key column names from the student table (marks uses sid, department uses dept_id).<br><br>

**Sample Data & Inputs:** <br>

import pandas as pd<br>

## Table 1: Student registration<br>
students = pd.DataFrame({<br>
    "student_id": [1, 2, 3, 4, 5],<br>
    "name":       ["Student A", "Student B", <br>"Student C", "Student D", "Student E"],<br>
    "dept_id":    [10, 20, 10, 30, 20]<br>
})

## Table 2: Marks (key column is named 'sid', not 'student_id')
marks = pd.DataFrame({<br>
    "sid":   [1, 2, 3, 5, 6],<br>
    "marks": [92, 75, 88, 61, 95]<br>
})

## Table 3: Department (key column is 'dept_id')
departments = pd.DataFrame({
    "dept_id":   [10, 20, 30],<br>
    "dept_name": ["Computer Science", <br>"Mechanical", "Electrical"]<br>
})
 <br><br>

**Expected output for the sample above:** <br>

After the inner join (students + marks on mismatched keys), left join with departments, filtering marks > 80, and groupby average:<br>
 <br>
dept_name  marks<br>
Computer Science	90.0

```python
departments = pd.DataFrame({ "dept_id": [10, 20, 30],
"dept_name": ["Computer Science",
"Mechanical", "Electrical"]
})
departments
```

```python
df = students.merge(marks, left_on='student_id', right_on='sid', how='inner').merge(departments, on='dept_id', how='left').drop('sid', axis=1).query('marks > 80').groupby('dept_name')['marks'].mean()
df
```

import pandas as pd

students = pd.DataFrame({
"student_id": [1, 2, 3, 4, 5],
"name": ["Student A", "Student B",
"Student C", "Student D", "Student E"],
"dept_id": [10, 20, 10, 30, 20]
})

marks = pd.DataFrame({
"sid": [1, 2, 3, 5, 6],
"marks": [92, 75, 88, 61, 95]
})

departments = pd.DataFrame({ "dept_id": [10, 20, 30],
"dept_name": ["Computer Science",
"Mechanical", "Electrical"]
})

df = students.merge(marks, left_on='student_id', right_on='sid', how='inner').merge(departments, on='dept_id', how='left').drop('sid', axis=1).query('marks > 80').groupby('dept_name')['marks'].mean()
