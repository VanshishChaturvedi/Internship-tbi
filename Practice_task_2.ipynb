````python
import numpy as np
import pandas as pd
````

````python
data = {
    'employee_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name':        ['Alice', 'Bob', 'Carol', 'Dave', 'Eve', 'Frank', 'Grace', 'Hank'],
    'department':  ['HR', 'it', None, ' Finance ', 'IT', None, 'hr', 'Finance'],
    'age':         [25.0, 32.0, None, 28.0, 30.0, None, 27.0, 35.0],
    'salary':      ['₹45,000', '₹60,000', None, '₹1,20,000', '₹50,000', 'Not Available', '₹55,000', '₹70,000'],
    'joining_date':['2022-01-15', '15/03/2021', 'invalid date', '2020-07-01', '10/08/2019', None, '2023-05-20', '2021-11-30'],
    'city':        ['Mumbai', None, 'Delhi', 'Bangalore', None, 'Chennai', 'Pune', 'Hyderabad']
}

df = pd.DataFrame(data)
````

````python
df
````

````python
df.isna().sum()
````

````python
df['department'] = df['department'].fillna('Unknown')
df['department'] = df['department'].astype(str).str.strip().str.title()
df
````

````python
df['salary'] = df['salary'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
df['salary'] = df['salary'].fillna(np.median(df['salary'].dropna()))
df
````

````python
df['joining_date'] = df['joining_date'].astype(str).str.replace('/', '-', regex=False)
df['joining_date'] = pd.to_datetime(df['joining_date'], errors='coerce', dayfirst=True, format='mixed')
df['joining_date'] = df['joining_date'].fillna(pd.to_datetime('1900-01-01'))
df
````

````python
df['age'] = df['age'].astype(float)
df['age'] = df['age'].fillna(df['age'].median())
df['age'] = df['age'].astype(int)
df
````

````python
df = df.fillna("Unkonwn")
df
````

````python
df.isnull().sum()
````
