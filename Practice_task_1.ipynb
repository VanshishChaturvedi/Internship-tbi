#Problem Statement:
A data analyst at an e-learning company is onboarding a raw dataset of student exam records. The dataset has data quality issues: some rows have missing marks or missing city entries. The analyst must write a Python function using Pandas that:
* Builds the DataFrame from the inlined data dictionary.
Inspects it with info() and describe().
Counts missing values per column using isnull().sum().
Filters rows where Marks > 75 using Boolean indexing.
Returns a summary dictionary containing:
* "filtered_shape"
* "columns"
* "total_missing"     <br><br>
**Expected Output:**  <br><br>
{   
'filtered_shape': (4, 3),  
'columns': ['Name', 'Marks', 'City'],  
'total_missing': 4   
}


Importing the required libraries

```python
import numpy as np
import pandas as pd
```

Building the function

```python
def summarise_records():
  data = {
      "Name": ["Student1", "Student2", "Student3", "Student4", "Student5", "Student6", "Student7"],
      "Marks": [52,85,np.nan,95,85,np.nan,76],
      "City": ["City1", "City2", "City3", "City4", "City5", "City6", "City7"]
  }
  df = pd.DataFrame(data)
  print("-"*40)
  print("INFO")
  display(df.info())

  print("-"*40)
  print("DESCRIBE")
  display(df.describe())

  print("-"*40)
  print("Number of missing values per column")
  display(df.isnull().sum())

  print("-"*40)
  print("Students Scoring more than 75 marks")
  filtered_df = df[df["Marks"] > 75]
  display(filtered_df)

  print("-"*40)
  summary = {
      "filtered_shape" : filtered_df.shape,
      "columns": list(df.columns),
      "total_missing" : df.isnull().sum().sum()
  }
  return summary
```

Calling the Function

```python
summarise_records()
```
